```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# We create a scatter plot to visualize the first two features (sepal length and sepal width) of the Iris dataset. Different colors represent different species.
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50)
plt.title("Iris Dataset (First Two Features)")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
# Get the cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
df=pd.DataFrame(X)
df['clusterLabel']=labels
df
```

```python
kmeans.predict([[7.1, 3.5 , 4.6, 1.5]])
```

```python
# We create a scatter plot to visualize the clusters assigned by K-Means. The centroids of each cluster are marked with red 'X' symbols. The legend indicates the position of the centroids.
plt.scatter(X[:, 2], X[:, 3], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 2], centroids[:, 3], marker='X', s=200, linewidths=3, color='red', label='Centroids')
plt.title("K-Means Clustering on Iris Dataset")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend()
plt.show()
```
