# KNN Classification

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df=pd.DataFrame({"x":[1,2,3,4,6,8],"y":[4,3,5,15,12.5,21],"class":[0,0,0,1,1,1]})
df
```

```python
import seaborn as sns
sns.scatterplot(data=df, x="x", y="y",hue="class")
```

```python
from sklearn.neighbors import KNeighborsClassifier
X=df[['x','y']]
y=df["class"]
# We're skipping the step of splitting the data for simplicity.
kn = KNeighborsClassifier(n_neighbors=3)
model=kn.fit(X, y)
```

```python
model.predict([[4,12]])
```

# KNN Regression

```python
import pandas as pd
df=pd.DataFrame({"x":[1,2,3,4,5,8],"y":[4,3,5,15,12,21],"target":[100,200,50,110,160,30]})
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='x', y='y', legend=False, sizes=(100, 200))
plt.xlim([0,9]);plt.ylim([0,25])
# Annotating each point with its target value
for i, txt in enumerate(df['target']):
    plt.annotate(txt, (df['x'][i] + 0.1, df['y'][i]))  # Adjusted text position for clarity

plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
X=df[['x','y']]
y=df["target"]
# We're skipping the step of splitting the data for simplicity.
kn = KNeighborsRegressor(n_neighbors=3)
model=kn.fit(X, y)
```

```python
print(model.predict([[4,10]]))
```
