## Convex Function

A convex function is a function where the line segment joining any two points on the graph of the function lies above or on the graph itself. In other words, if you pick any two points on the graph of a convex function and draw a straight line between them, the line segment will not dip below the graph at any point. Convex functions have a **single global minimum**, meaning that there's only one point in the function's domain where it achieves its lowest value.

image.png

In the case of the least squares function, the function typically looks like:

$$ \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

after replacing ${\hat{y_i}}$ with $mx + b$  we get,  
  $$\sum_{i=1}^{N} (y_i - mx_i - b)^2$$


This function is a convex function. The **domain** of the function is the set of all possible values for the parameters and the **range** of the function is the cost for different parameter values.


### Convexity Proof

Let's delve deeper into why the square term $ (y_i - \hat{y}_i)^2 $ is convex.

A function $ f(x) $ is considered convex if its second derivative is non-negative over its entire domain. We can illustrate this property for the square function $ f(x) = x^2 $:

1. **First Derivative Test**: The first derivative of $ f(x) = x^2 $ is $ f'(x) = 2x $. This derivative represents the slope of the function at any point $ x $.

2. **Second Derivative Test**: The second derivative of $ f(x) $ is $ f''(x) = 2 $, which is a constant value. This means that the second derivative is positive everywhere, indicating that the function is concave up over its entire domain.

3. **Convexity Implication**: Functions that are concave up (i.e., where the second derivative is positive) are convex. This is because the slope is constantly increasing as we move along the function from left to right.

Now, let's apply this understanding to the term $ (y_i - \hat{y}_i)^2 $ in the context of the least squares function:

- $ (y_i - \hat{y}_i)^2 $ represents the squared difference between the observed value $ y_i $ and the predicted value $ \hat{y}_i $.
  
- Squaring this difference ensures that the result is always non-negative, as the square of any real number is non-negative.

- Additionally, the function $ (y_i - \hat{y}_i)^2 $ is symmetric around the point where $ y_i = \hat{y}_i $, meaning it curves upward on both sides of this point.

- This upward curvature (confirmed by the second derivative test) indicates that the function is convex, as the slope is constantly increasing over its entire domain.

Thus, the term $ (y_i - \hat{y}_i)^2 $ is convex, contributing to the overall convexity of the least squares function.

## Dervative, Partial Derivative, and Gradient

<h3> Derivative

A derivative is a fundamental concept in calculus that represents the rate of change of a function at a particular point. It provides information about how a function's value changes as its input (independent variable) changes.


<h3> Partial Derivative

A partial derivative is a concept from calculus that is used to determine the rate of change of a function when the function has more than one input. It measures how a function changes when you vary one input variable, treating all the other input variables as if they were fixed.

Let's consider a function of multiple variables as **f(x, y, z, ...)**, and you want to find the partial derivative of this function with respect to one of the variables, say **x**, you would denote it as **f/x** (read as "partial f partial x"). The partial derivative with respect to **x** is computed by treating all other variables, like **y, z, etc.**, as constants and applying the standard rules of differentiation for single-variable functions.

For example, if you have a function $f(x, y) = x^2 + 2xy + y^2$, and you want to find **f/x**, you would differentiate the function with respect to **x** while treating y as a constant. This would give you **f/x = 2x + 2y**.

<h3> Gradient

In mathematics and particularly in the context of optimization and machine learning, the gradient is a vector that points in the direction of the steepest increase of a function at a specific point. It provides information about how the function's output changes as you move away from that point. More formally, the gradient of a function is a vector composed of its partial derivatives with respect to each of its input variables.

For a function of multiple variables, such as $f(x, y, z)$, the gradient is represented as the vector $\nabla f$, and its components are:

$
\nabla f = \left[ \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z} \right]$

Each component of the gradient vector indicates how the function changes with respect to each input variable.

**The gradient points in the direction of the steepest increase, so moving in the opposite direction (i.e., descending along the negative gradient) is a way to find the minimum of the function.**

## Optimization

Optimization in machine learning refers to the process of finding the best possible set of parameters (**slope**) for a model that minimizes or maximizes the **cost function/objective function**. The goal of optimization is to adjust the model's parameters in such a way that the cost is minimized.

<h3>Key components of optimization in machine learning:

1. **Cost Function:** Define a mathematical function that quantifies the cost of the model. This function is typically based on the difference between the model's predictions and the actual target values.
  
  $$\sum_{i=1}^{N} (y_i - mx_i - b)^2$$


2. **Model Parameters:** Machine learning models have parameters that need to be tuned to minimize the cost function. These parameters include **slope** and **y intercept** in simple linear regression, weights and biases in neural networks, or other model-specific parameters.

3. **Optimization Algorithm:** Choose an optimization algorithm that will iteratively update the model's parameters to find the ones that minimize the cost function. Common optimization algorithms include **Gradient Descent**, **Adam**, **RMSprop**, and more. ***These algorithms adjust the parameters in the direction that reduces the loss.***

## Gradient Descent Algorithm

Gradient descent is an optimization algorithm used to minimize the cost function in machine learning and deep learning models during the training process. The algorithm finds the minimum of a function by iteratively moving in the direction of the steepest descent as defined by the negative gradient of the function.

<h3>Here's how it works:

1. **Initialize parameter**: Starts with an initial guess for the parameters of the model.

2. **Compute Gradient**: Computes the gradient of the cost function with respect to the parameter. The gradient indicates the direction of steepest increase of the function at the current parameter.

3. **Update parameter**: Adjusts the parameter parameter in the direction opposite to the gradient using the formula:

   $ \text{New Parameter} = \text{Old Parameter} - \alpha \times \text{Gradient} $

   where:
   - $\alpha$ (alpha) is the learning rate. Common values include 0.1, 0.01, 0.001, etc.
   - $\text{Gradient}$ is the gradient of the cost function with respect to the parameter parameter.

4. **Check Convergence Criteria**: Repeats steps 2 and 3 until a stopping criterion is met.

5. **Output**: Once the algorithm stops, the final parameter represents the solution that minimizes the cost function (or is close to the minimum).

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2

def f_prime(x):
    return 2*x

# Gradient descent parameters
learning_rate = 0.3
n_iterations = 10
initial_x = -7.5  # Starting point

# Gradient descent algorithm
x = initial_x
x_history = [x]

for iteration in range(n_iterations):
    gradient = f_prime(x)
    x = x - learning_rate * gradient
    x_history.append(x)

# Plot the results
x_values = np.linspace(-10, 10, 400)
y_values = f(x_values)

plt.plot(x_values, y_values, label='$y = x^2$')
plt.scatter(x_history, [f(x) for x in x_history], color='red')
plt.plot(x_history, [f(x) for x in x_history], 'ro-', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent on $y = x^2$')
plt.legend()
plt.show()

print("Final value of x after gradient descent:")
print(x)
```


<h3>Explanation</h3>

1. **Function and Derivative**: We define the function $ f(x) = x^2 $ and its derivative $ f'(x) = 2x $.
2. **Parameters**:
    - `learning_rate`: Controls the size of the steps taken during gradient descent.
    - `n_iterations`: The number of iterations to perform gradient descent.
    - `initial_x`: The starting point for gradient descent.
3. **Gradient Descent**:
    - Initialize $ x $ to the starting point.
    - In each iteration, compute the gradient $ f'(x) $.
    - Update $ x $ using the learning rate and the gradient.
    - Store the history of $ x $ values to visualize the path of gradient descent.
